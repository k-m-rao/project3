```python
import PIL
from PIL import Image
from tkinter.filedialog import *
import zlib, base64,sys, timeit
import ffmpeg, subprocess, sys

def compress_text_file():
    text_file_path = askopenfilename()
    text = open(text_file_path, 'r').read()
    print('Raw Size : ', sys.getsizeof(text))
    #startt = time.time()
    startt = timeit.default_timer()
    compressed_text = base64.b64encode(zlib.compress(text.encode(), 1))

    print('Time taken to compress:', timeit.default_timer() - startt)
    print('compressed_text:', compressed_text)
    print('compressed Size : ', sys.getsizeof(compressed_text))
    save_path = asksaveasfilename()
    f = open(save_path + '.txt', 'w')
    f.write(compressed_text.decode())
    print('SUCCESS\n')
    f.close()

def decompress_text_file():
    compressed_file = askopenfilename()
    f = open(compressed_file, 'r')
    compressed_text = f.read()
    decompressed_text = zlib.decompress(base64.b64decode(compressed_text))
    f.close()
    print('decompressed_text: ',decompressed_text)
    print('SUCCESS\n')

def compress_image():
    try:
        file_path = askopenfilename()
        img = PIL.Image.open(file_path)
        myHeight, myWidth = img.size
        img = img.resize((myHeight, myWidth), PIL.Image.ANTIALIAS)
        save_path = asksaveasfilename()
        img.save(save_path+'_compressed.JPG')
        print('SUCCESS\n')
    except Exception as E:
        print(E)        

def compress_video():
    sample_ip = askopenfilename()
    s1 = sys.getsizeof(sample_ip)
    print('Raw Size : ', s1)
    
    subprocess.run('ffmpeg -i ' + str(sample_ip) + ' -b 45k sample_video_output.mp4')
    s2 = sys.getsizeof('sample_video_output.mp4')
    print('compressed Size : ', s2)   
    print('SUCCESS\n')
    
def compress_audio():
    sample_ip = askopenfilename()
    s1 = sys.getsizeof(sample_ip)
    print('Raw Size : ', s1)    
    
    cmd = 'ffmpeg -i ' + str(sample_ip) + ' -ab 96k sample_audio_output.mp3'
    result = subprocess.run(cmd)
        
    s2 = sys.getsizeof('sample_audio_output.mp3')
    print('compressed Size : ', s2)
    print('SUCCESS\n')
    
def compress_pdf(power=0):
    #compression levels from least to highest
    file_to_compress = askopenfilename()
    output_file = 'sample_pdf_output.pdf'
    quality = {
        0: '/default',
        1: '/prepress',
        2: '/printer',
        3: '/ebook',
        4: '/screen' }
    print("Compress PDF...")
    initial_size = os.path.getsize(file_to_compress)

    
    script = "gswin64c -dNOPAUSE -dBATCH -sDEVICE=pdfwrite -dCompatibilityLevel=1.4 -dPDFSETTINGS={} -sNAME=setting -sOutputFile={} {}".format(
        quality[power], output_file, file_to_compress)
    command = script.split(" ")
    subprocess.run(command)

    final_size = os.path.getsize(output_file)
    ratio = 1 - (final_size / initial_size)
    print("Initial file size is {}KB".format(round(initial_size/1000)))
    print("Final file size is {}KB".format(round(final_size/1000)))
    print("Compression by {0:.0%}.".format(ratio))
    print('SUCCESS\n')

repeat = 1
while(repeat):
    print("\n\n1: Compress Image")
    print("2: Compress .txt File")
    print("3: Decompress .txt compressed File")
    print('4: Compress .mp4 video file')
    print('5: Compress .mp3 Audio file')
    print('6: Compress PDF file')
        
    choice = int(input("Enter your choice"))
    if choice == 1:
        print("You have chosen to Compress Image")
        print("Select the source and destination in the new window shown:")
        compress_image()
    elif choice == 2:
        print("You have chosen to Compress .txt file")
        print("Select the source and destination in the new window shown:")
        compress_text_file()
    elif choice == 3:
        print("You have chosen to Decompress .txt file")
        print("Select the source file in the new window shown:")
        decompress_text_file()
    elif choice == 4:
        print("You have chosen to Compress .mp4 video file")
        compress_video()
    elif choice ==5:
        print("You have chosen to compress .mp3 audio file")
        compress_audio()
    elif choice == 6:
        compress_pdf(power=4)
    else:
        print("Please enter correct option")
        
    repeat = int(input('press 1 to continue 0 to quit'))
    
```
